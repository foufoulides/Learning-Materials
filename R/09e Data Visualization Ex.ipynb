**Histogram of hwy mpg values**

```r
pl1 <- ggplot(data=mpg, aes(x = hwy)) + 
    geom_histogram(binwidth=2, fill="lightblue", color="lightgrey", alpha=0.5)

pl1
```

**Barplot of car counts per manufacturer with color fill defined by cyl count**

```r
pl2 <- ggplot(data=mpg, aes(x=factor(manufacturer))) + 
        geom_bar(aes(fill=factor(cyl)))

pl2
```

**Scatterplot of volume vs sales**
* Play around with **`alpha`** to clarify the information.

```r
pl3 <- ggplot(data=txhousing, aes(x=sales, y=volume)) + 
        geom_point(alpha=0.5, color="blue")
pl3
```

**Add a smooth line to the scatterplot above**

```r
pl3 <- ggplot(data=txhousing, aes(x=sales, y=volume)) + 
        geom_point(alpha=0.7, color="lightblue") + 
        stat_smooth(color="red")
        
pl3
```
