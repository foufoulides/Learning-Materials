## <center> Team Organogram with NetworkX </center> 

```python
import networkx as nx 
import matplotlib.pyplot as plt
import PIL
```

```python
image = PIL.Image.open("PersonIcon.JPG")
```

```python
Prisons = nx.Graph()
```

```python
Prisons.add_edges_from([
    ("Olivia Portis", "Will Miller"), 
    ("Olivia Portis", "Ross Kennedy"), 
    ("Olivia Portis", "Lucy Pitches"), 
    ("Will Miller", "Laura Tracey"), 
    ("Will Miller", "Tim Culwick"), 
    ("Will Miller", "Holly Gatward"),
    ("Ross Kennedy", "Chris Foufoulides"),
    ("Ross Kennedy", "Vacancy 1"),
    ("Ross Kennedy", "Vacancy 2"),
    ("Lucy Pitches", "Ryan OMissenden"),
    ("Lucy Pitches", "Vacancy 3"), 
    ("Lucy Pitches", "Vacancy 4")])
```

```python
nx.draw(Prisons, with_labels=True, edge_color="grey", alpha=0.8)
```

```python
nx.draw(Prisons, pos=nx.nx_agraph.graphviz_layout(Prisons, prog="twopi", args=""),
                 with_labels=True, alpha=0.8, edge_color="grey", horizontalalignment="center", verticalalignment="top",
                 font_weight="normal", font_size=12, font_color="black", node_size=1400)
```
